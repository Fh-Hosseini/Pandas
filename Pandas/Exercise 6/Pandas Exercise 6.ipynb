```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('yektanet.csv')
df.tail()
```

<h2 align=left style="line-height:200%;color:#0099cc">
<font color="#0099cc">
Part One
</font>
</h2>

<p dir=ltr style="direction: ltr;text-align: justify;line-height:200%;font-size:medium">
<font size=3>
Using grouping and the filter function, select websites whose keywords appear more than 1400 times in the dataset. In other words, the list of keywords should be repeated more than 1400 times in the dataframe. Save the result in the variable 'filtered_df'.

Guide: First apply the groupby function for keywords, and then use the len function to check the number of occurrences. Note that it is not necessary to group based on each element inside the array of keywords.
</font>
</p>

```python
filtered_df = df.groupby('keywords').filter(lambda x: len(x)>1400)
filtered_df
```

<h2 align=left style="line-height:200%;color:#0099cc">
<font color="#0099cc">
Part Two
</font>
</h2>

<p dir=ltr style="direction: ltr;text-align: justify;line-height:200%;font-size:medium">
<font size=3>
Now add a column named 'popularity' to the 'filtered_df' dataframe, which includes the ratio of the number of visits to each website to the total number of visits for all websites with that keyword.

Guide: After grouping by keywords, apply the transform function on the 'visit count' column and calculate the ratio of visits for each website to the total visits of that keyword group. Multiply the result by 100.
</font>
</p>

```python
filtered_df['popularity']  = filtered_df.groupby('keywords')['visit count'].transform(lambda x: x/x.sum()*100)
filtered_df
```
