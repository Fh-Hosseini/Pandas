<h2 align=left style="line-height:200%;color:#0099cc">
<font color="#0099cc">
Dataset
</font>
</h2>

<p dir=ltr style="direction: ltr;text-align: justify;line-height:200%;font-size:medium">
<font size=3>
Book information is available in the 'books_db.csv' file, authors in the 'authors.csv' file, publishers in the 'publisher.csv' file, and languages in the 'language.csv' file. Using pandas, read each of these files and store them in the corresponding variables.
</font>
</p>

```python
import pandas as pd
import numpy as np

df = pd.read_csv("books_db.csv")
authors = pd.read_csv("authors.csv")
language = pd.read_csv("language.csv")
publisher = pd.read_csv("publisher.csv")

df.head()
```

<h2 align=left style="line-height:200%;color:#0099cc">
<font color="#0099cc">
Part One
</font>
</h2>

<p dir=ltr style="direction: ltr;text-align: justify;line-height:200%;font-size:medium">
<font size=3>
As you can see in the main dataset (books), the values in the three columns 'lang_id,' 'author_id,' and 'publisher_id' are only in the form of IDs. These IDs correspond to the index numbers in the other dataframes. In this section, we intend to extract their complete information from the respective dataframes according to these indices and add them to the df dataframe. Therefore, merge the three dataframes 'authors,' 'language,' and 'publisher' appropriately with the df dataframe.
</font>
</p>

```python
df = pd.merge(df, authors, left_on="author_id", right_index=True)
df = pd.merge(df, language, left_on="lang_id", right_index=True)
df = pd.merge(df, publisher, left_on="publisher_id", right_index=True)

df.head()
```

<p dir=ltr style="direction: ltr;text-align: justify;line-height:200%;font-size:medium">
<font size=3>
Now that all the values have been added, we no longer need the three 'id' columns that we used as keys. Therefore, we will use the 'drop' command to remove them.    
</font>
</p>

```python
df.drop(columns = ['author_id', 'lang_id', 'publisher_id'], inplace = True)
```

<h2 align=left style="line-height:200%;color:#0099cc">
<font  color="#0099cc">
Part Two
</font>
</h2>

<p dir=ltr style="direction: ltr;text-align: justify;line-height:200%;font-size:medium">
<font size=3>
Now, change the column names of the dataframe in a way that, from left to right, they correspond to the following names:
<ul>
<li><code>bookID</code></li>
<li><code>title</code></li>
<li><code>average_rating</code></li>
<li><code>isbn</code></li>
<li><code>isbn13</code></li>
<li><code>num_pages</code></li>
<li><code>ratings_count</code></li>
<li><code>text_reviews_count</code></li>
<li><code>publication_date</code></li>
<li><code>author</code></li>
<li><code>lang</code></li>
<li><code>publisher</code></li>
</font>
</p>

```python
df.rename( columns = { 'name_x': 'author', 'name_y': 'publisher', 'code': 'lang'}, inplace=True)
```

<h2 align=left style="line-height:200%;color:#0099cc">
<font color="#0099cc">
Part Three
</font>
</h2>

<p dir=ltr style="direction: ltr;text-align: justify;line-height:200%;font-size:medium">
<font size=3>
Now, calculate the average rating of books from each publisher and store the result in the variable 'publisher_rating'.
</font>
</p>

```python
publisher_rating = df.groupby("publisher")['average_rating'].mean()
publisher_rating
```

<h2 align=left style="line-height:200%;color:#0099cc">
<font  color="#0099cc">
Part Four
</font>
</h2>

<p dir=ltr style="direction: ltr;text-align: justify;line-height:200%;font-size:medium">
<font  size=3>
In this section, first add a column named 'num_pages_grp' to the dataframe, where the number of pages in the book is represented in the following ranges.
</font>
</p>

<center>
<div dir=ltr style="direction: ltr;line-height:200%;font-size:medium">
<font size=3>

| <b>Value</b> | <b>Description</b> |
| :---: | :---: |
| <code dir=ltr><250</code> | Less than 250 pages|
| <code dir=ltr>250-499</code> | 250 to 499 pages|
| <code dir=ltr>500-749</code> | 500 to 749 pages|
| <code dir=ltr>750-999</code> | 750 to 999 pages|
| <code dir=ltr>>=1000</code> | 1000 pages or more|

</font>
</div>
</center>


<p dir=ltr style="direction: ltr;text-align: justify;line-height:200%;font-size:medium">
<font size=3>
For this task, you can use the np.select function.
</font>
</p>


```python
x = df['num_pages']

conditions = [x<250, x<500, x<750, x<1000, x>=1000]
options = ["<250", "250-499", "500-749", "750-999", ">=1000"]
df["num_pages_grp"] = np.select(conditions, options)
df.head()
```


<p dir=ltr style="direction: ltr;text-align: justify;line-height:200%;font-size:medium">
<font size=3>
Now we want to know how many books each author has written in each of these ranges.

</font>
</p>

```python
author_page = df.groupby(['author', 'num_pages_grp']).size()
author_page
```
