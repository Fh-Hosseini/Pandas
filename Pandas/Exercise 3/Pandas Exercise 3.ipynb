<h2 align=left style="line-height:200%;color:#0099cc">
<font color="#0099cc">
Part One
</font>
</h2>

<p dir=ltr style="direction: ltr;text-align: justify;line-height:200%;font-size:medium">
<font size=3>
In this social network, there are 5 categories of being a branch. To be a member of each category, our page and name must meet specific conditions.
The first category of branches in this network is called Berenji. Berenji branches consist of individuals whose number of followers (following) by their page is less than 11.
In the cell below, separate a portion of the main DataFrame that meets the conditions for Berenji branches and store it in the variable 'berenji_df'.
</font>
</p>

```python
berenji_df = df[df.following < 11]
berenji_df
```

<h2 align=left style="line-height:200%;color:#0099cc">
<font color="#0099cc">
Part Two
</font>
</h2>

<p dir=ltr style="direction: ltr;text-align: justify;line-height:200%;font-size:medium">
<font size=3>
In the variable 'mesi_df', store Mesi branches. Mesi branches consist of individuals whose number of posts on their page is greater than the number of individuals following that page.
</font>
</p>

```python
mesi_df = df[df.post_number > df.following]
```

<h2 align=left style="line-height:200%;color:#0099cc">
<font color="#0099cc">
Part Three
</font>
</h2>

<p dir=ltr style="direction: ltr;text-align: justify;line-height:200%;font-size:medium">
<font size=3>
In the variable 'boronzi_df', store Bronze branches. Bronze branches consist of individuals whose number of followers (follower) on their page is greater than one million.
</font>
</p>

```python
boronzi_df = df[df.follower > 1000000]
```

<h2 align=left style="line-height:200%;color:#0099cc">
<font  color="#0099cc">
Part Four
</font>
</h2>

<p dir=ltr style="direction: ltr;text-align: justify;line-height:200%;font-size:medium">
<font size=3>
In the variable 'noghreyee_df', store Silver branches. Silver branches consist of individuals for whom both of the following conditions are simultaneously satisfied:
    <ol dir=ltr>
        <li>The number of followers (follower) on their page is greater than or equal to 10 million.</li>
        <li>The number of individuals followed by their page (following) is less than 50.</li>
    </ol>
</font>
</p>

```python
noghreyee_df = df[ (df.follower >= 10000000) & (df.following < 50)]
```

<h2 align=left style="line-height:200%;color:#0099cc">
<font color="#0099cc">
Part Five</font>
</h2>

<p dir=ltr style="direction: ltr;text-align: justify;line-height:200%;font-size:medium">
<font size=3>
In the variable 'talayee_df', store Golden branches. Golden branches consist of individuals who have at least one of the following two conditions (meaning either the first condition or the second one):
<ol dir=ltr>
        <li>The number of followers (follower) on their page is greater than 10 million, and the number of individuals followed by them (following) is less than 20.</li>
        <li>The number of their posts is less than 15, and the letter 'R' exists in the owner's page name (e.g., !BooAzaR).</li>
    </ol>
</font>
</p>

```python
talayee_df = df[ (df.follower > 10000000 & df.following < 20) || (df.post_number < 15)]
```
