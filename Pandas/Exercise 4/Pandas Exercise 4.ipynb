```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('food.csv')
df.head()
```

<h2 align=left style="line-height:200%;color:#0099cc">
<font color="#0099cc">
Part One
</font>
</h2>
<p dir=ltr style="direction: ltr;text-align: justify;line-height:200%;font-size:medium">
<font size=3>
We have often heard that Indians love spicy foods! Now, let's investigate the accuracy of this statement based on the available dataset and see what proportion of the total foods in this dataset are spicy. To do this, simply calculate the ratio of the number of spicy foods to the total number of foods and store it in the variable 'spicy_ratio'.
</font>
</p>


```python
spicy_ratio = len(df[df['flavor_profile'] == 'spicy']) / len(df)
spicy_ratio
```

<h2 align=left style="line-height:200%;color:#0099cc">
<font color="#0099cc">
Part Two
</font>
</h2>
<p dir=ltr style="direction: ltr;text-align: justify;line-height:200%;font-size:medium">
<font size=3>
Now, we want to know which foods take the longest time from the start of preparation to being ready to serve. For this, you need to consider the total time spent in preparation and cooking, then store the top 10 foods with the longest preparation time in descending order in a DataFrame variable named 'df_most_time_consuming'.

Note: The format and order of columns in the 'df_most_time_consuming' DataFrame should be exactly similar to the 'df' DataFrame. This means that the DataFrame will have 10 rows and 9 columns.
</font>
</p>


```python
df_most_time_consuming = df.iloc[df.apply(lambda row: row['prep_time'] + row['cook_time'], axis=1).argsort()[:-11:-1]]
df_most_time_consuming
```
