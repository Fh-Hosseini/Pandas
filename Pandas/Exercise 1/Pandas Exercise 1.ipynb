<h2 align=left style="line-height:200%;color:#0099cc">
<font color="#0099cc">
Part One: Ideal Segment
</font>
</h2>
<p dir=ltr style="direction: ltr;text-align: justify;line-height:200%;font-size:medium">
<font  size=3>
To assess the quality of diamond cuts, we sought the assistance of an expert. They documented the quality of each diamond in a list and provided us with the following final list:
</font>
</p>


```python
cut = ['Ideal', 'Premium', 'Premium', 'Premium', 'Premium', 'Very Good',
       'Premium', 'Premium', 'Ideal', 'Premium', 'Good', 'Good', 'Premium',
       'Very Good', 'Very Good', 'Good', 'Good','Very Good', 'Good', 'Good',
       'Premium', 'Premium', 'Good', 'Good']
```

<p dir=ltr style="direction: ltr;text-align: justify;line-height:200%;font-size:medium">
<font size=3>
Using this information, create a pandas Series named "cut_series."
</font>
</p>

```python
import pandas as pd
cut_series = pd.Series(cut)
```

<p dir=ltr style="direction: ltr;text-align: justify;line-height:200%;font-size:medium">
<font size=3>
In the "cut_series", the quality of the cut for each diamond is stored. Now, store the count of each cut quality in ascending order in a new Series named "cut_sorted." This means the title and count of the least frequent cut type should be at the beginning of this Series.
</font>
</p>

```python
cut_sorted = cut_series.value_counts(ascending=True)
cut_sorted
```

<h2 align=left style="line-height:200%;color:#0099cc">
<font color="#0099cc">
Part Two: Diamond Carats
</font>
</h2>
<p dir=ltr style="direction: ltr;text-align: left;line-height:200%;font-size:medium">
<font size=3>
Using a scale, we weighed the diamonds and recorded the results in the following list.
</font>
</p>


```python
carat = [1.07, 1.19, 1.19, 1.2, 1.2, 1.25, 1.2, 1.2, 1.2,
         1.14, 1.01, 1.01, 1.01, 1.02, 1.01, 1.01, 1.01,
         1.01, 1.01, 1.01, 1.01, 1.01, 1.0, 1.0]
```

<p dir=ltr style="direction: ltr;text-align: justify;line-height:200%;font-size:medium">
<font size=3>
Using this information, create a pandas Series named "carat_series."
</font>
</p>


```python
carat_series = pd.Series(carat)
```

<p dir=ltr style="direction: ltr;text-align: justify;line-height:200%;font-size:medium">
<font size=3>
Calculate the 75th percentile of this Series and store it in the variable "carat_75."
</font>
</p>


```python
import numpy as np
carat_75 = np.percentile(carat_series, 75)
```

<h2 align=left style="line-height:200%;color:#0099cc">
<font  color="#0099cc">
Part Three: Diamond Dimensions
</font>
</h2>
<p dir=ltr style="direction: ltr;text-align: justify;line-height:200%;font-size:medium">
<font size=3>
At this stage, we asked another expert to measure the diamonds, and they provided us with a NumPy array. One of the ways to store a NumPy array is to use the npz storage format. In this file format, you can save any number of NumPy arrays, and later, by using the array names, extract the desired array from the file.
</font>
</p>


```python
import numpy as np

sizes = np.load('sizes.npz', allow_pickle=True)
sizes_array = sizes['sizes']
sizes_array
```


<p dir=ltr style="direction: ltr;text-align: left;line-height:200%;font-size:medium">
<font size=3>
The `sizes_array` has three columns, where the first column represents length, the second column represents width, and the third column represents the height of the diamonds. Using this array, create a DataFrame named `sizes_df` and assign the column names as x, y, and z (with the first column named x).
 </font>
</p>


```python
sizes_df = pd.DataFrame(sizes_array, columns=['x', 'y', 'z'])
sizes_df.head()
```

<h2 align=left style="line-height:200%;color:#0099cc">
<font color="#0099cc">
Part Four: Indexing
</font>
</h2>
<p dir=ltr style="direction: ltr;text-align: justify;line-height:200%;font-size:medium">
<font size=3>
Contrary to the programming world where numbering usually starts from 0, in the diamond business, it is customary to start indexing from 1. Therefore, modify the indices accordingly so that they start from 1. Save the new DataFrame in `sizes_new_index`.

</font>
</p>


```python
sizes_new_index = sizes_df.copy()
sizes_new_index.index = np.arange(1, len(sizes_new_index)+1)
sizes_new_index
```
